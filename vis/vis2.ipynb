```python
# Adding Maplotlib module
import matplotlib.pyplot as plt
```

```python
x = [i for i in range(5)]
y = [i for i in range(5)]
m = ['.', 'o', 'v', '^', '<']
col = ['r', 'g', 'b', 'c', 'y']
for i, j, k, l in zip(x, y, m, col):
  plt.scatter(i, j, marker=k, c=l)
```

```python
from math import sin
x = [i for i in range(-5, 5)]
y = [sin(i) for i in x]

for i, j in zip(x, y):
    txt = "{:.2f}, {:.0f}".format(i+0.05, j+0.05)
    plt.text(i+0.05, j+0.05, txt)
plt.scatter(x, y)
plt.show()
```
