```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

# Generate some data
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create a 3D figure
fig = plt.figure()
ax = plt.axes(projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Add labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D plot')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np

# Define the saddle function
def saddle(x, y):
    return x**2 - y**2

# Generate data
x = np.linspace(-2, 2, 50)
y = np.linspace(-2, 2, 50)
X, Y = np.meshgrid(x, y)
Z = saddle(X, Y)

# Create a 3D figure
fig = plt.figure()
ax = plt.axes(projection='3d')

# Plot the saddle
ax.plot_surface(X, Y, Z, cmap='coolwarm')

# Add labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Saddle plot')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

def sigmoid(x, y):
    return 1 / (1 + np.exp(-(x + y)))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = sigmoid(X, Y)

ax.plot_surface(X, Y, Z, cmap='coolwarm')
ax.set_title('Sigmoid Function')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

def softmax(x, y):
    exp_x = np.exp(x)
    exp_y = np.exp(y)
    return exp_x / (exp_x + exp_y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = softmax(X, Y)

ax.plot_surface(X, Y, Z, cmap='coolwarm')
ax.set_title('Softmax Function')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

def relu(x, y):
    return np.maximum(0, x + y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = relu(X, Y)

ax.plot_surface(X, Y, Z, cmap='coolwarm')
ax.set_title('ReLU Function')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
