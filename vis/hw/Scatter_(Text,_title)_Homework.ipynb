# Scatter text, legend, grid

```python
import matplotlib.pyplot as plt
```

## Text in Matplotlib in plots


### Title of plot
10. Insert Title of your plot as "Point coordinates"(Chizmaning sarlavhasini "Point coordinates" deb nomlang)

```python
def plot_title(x,y):
  '''
  Parameters: x and y axis points;
  '''
  plt.title("Point coordinates", c='r')
  plt.scatter(x, y)
  plt.show()

plot_title(3.0, 4.0)
```

### Text in the plot
11. Print text(Full Name) in the plot (chizmada matnni(to'liq ismingizni) chop eting)

```python
def text_in_plot(x,y):
  # plt.plot(9, 10)
  # Add text at (x, y) and color
  plt.text(x, y, 'World', color='red')

text_in_plot(0.5, 0.5)
```

*12*. Inserting coordinates for each points (har bir nuqta uchun kordinatalarni kiriting)

```python
def each_point_style_size_color_cor(x,y,color,style,size):
  '''
  Parameters: x and y axis points list;
  '''
  # Write the code here
  for i, j, col, st, siz in zip(x, y, color, style, size):
    plt.scatter(i, j, c = col, marker = st, s = siz)
    
  plt.show()

each_point_style_size_color_cor([1, 2, -1, 0], [2, 3, 6, 7], ['r', 'b', 'y', 'g'], ['o', '^', '<', '>'], [50, 60, 70, 40])
```

### Name of Axis
13. Insert name of Axises (O'qlarning nomini kiriting)

```python
def x_y_label(x,y):
  '''
  Parameters: x and y axis points;
  '''
  # Write the code here
  plt.scatter(x, y, c = 'r')
  plt.xlabel('X - axis', c = 'g')
  plt.ylabel('Y - axis', c = 'g')
  plt.show()

x_y_label(1, 2)
```

### Legend
14. Adding legend(description of data) for your plot (Chizmangizga legend (ma'lumotlar tavsifi) qo'shing)

```python
def legend_of_data(x,y):
  '''
  Parameters: x and y axis points list;
  '''
  # Write the code here
  colors = ['c', 'm', 'y']
  for i, j, col in zip(x, y, colors):
    plt.scatter(i, j, c = col, s = 50, label= f'Color {col}')

  plt.legend()
  plt.show()


legend_of_data([1, 2, 3], [3, 4, 2])
```

## Grid
15. Adding grid for your plot (Chizmangiz uchun grid(katak) qo'shing)

```python

def plot_grid(x,y,color,style,size):
  '''
  Parameters: x and y axis points list;
  '''
  # Write the code here
  plt.grid()
  for i, j, col, st, siz in zip(x, y, color, style, size):
    plt.scatter(i, j, c = col, marker = st, s = siz)
    
  plt.show()

plot_grid([1, 2, -1, 0], [2, 3, 6, 7], ['r', 'b', 'y', 'g'], ['+', '^', '<', '>'], [50, 60, 70, 40])
```
