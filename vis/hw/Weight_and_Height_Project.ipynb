### `weight-height.csv` fayilni yuklab oling.

 **Gender** - ustunini listga 0 va 1 ko'rinishda joylashtiring

 **Weight** - ustunini Pound da berilgan Kg ko'rinishda joylashtiring (1 kg = 2.205 pound).

 **Height** - ustunini inch da berilgan sm ko'rinishda listga joylashtiring (2.54 sm = 1 inch).

```python
# import library
import csv
import matplotlib.pyplot as plt
```

```python
def read_csv_data(file_path):
    """
    Read csv file
    Args:
        file_path(str): file path
    Returns:
        list: data split row
    """
    with open(file_path, 'r') as f:
        data = f.read().splitlines()[1:]
    
    return data

# read_csv_data('weight-height.csv')
```

```python
def get_data(data):
    """
    Get data from list.
    Gender: Change Male to 0 and Female to 1
    Weight: Place the column in the Kg view given in Pound (1 kg = 2.205 pound).
    Height: Place the column on the list in the cm view given in inches (2.54 cm = 1 inch).
    Args:
        data(list): data split row
    Returns:
        tuple: gender, weight and height

    """
    gender = []
    weight = []
    height = []
    # WRITE YOUR CODE HERE
    for d in data:
        d = d.split(',')
        gender.append(d[0])
        weight.append(float(d[1])*2.205)
        height.append(float(d[2])/2.54)

    return gender,weight,height

data = read_csv_data(file_path='weight-height.csv')
print(get_data(data)[0][0], get_data(data)[1][0], get_data(data)[2][0])
```

# Scatter plot
**-** **Weight** va **height** ma'lumotlarni **gender** orqali ajratib alohida listlarga joylashtiring.

**-** Ma'lumotlarni tasvirlang.

```python
plt.figure(dpi=300)

def show(gender, weight, height):
    """
    Show data.
    Args:
        gender(list): 0 and 1
        weight(list): weight
        height(list): height
    Returns:
        None
    """
    w_male = []
    w_female = []
    h_male = []
    h_female = []
    #WRITE YOUR CODE HERE
    for g, w, h in zip(gender, weight, height):
        if g == '"Male"':
            h_male.append(h)
            w_male.append(w)
        else:
            h_female.append(h)
            w_female.append(w)
    
    plt.scatter(w_female, h_female, c = 'r', marker='o', alpha=0.3, label= "Female")
    plt.ylabel("Weight")
    plt.xlabel("Height")
    plt.scatter(w_male, h_male, c = 'black', marker='o', alpha=0.3, label= "Male")

    plt.legend()
    plt.show()

data = read_csv_data('weight-height.csv')
gender, weight, height = get_data(data)

show(gender, weight, height)
```
